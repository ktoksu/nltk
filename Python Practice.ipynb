```python
#### File I/O

- reading a file:

f = open('text.txt')
for line in f:   OR   for line in f.readlines():
    print(line, end = '')

- writing to a file:

infile = open('C:\\Users\\kt\\Desktop\\lynda\\python\lines.txt', 'r')
outfile = open('C:\\Users\\kt\\Desktop\\new.txt', 'w')
for line in infile:
    print(line, file = outfile, end = '')
print('\nDone')

infile.close()
outfile.close()


#### urllib

- to scrape a webpage:

    import urllib.request
    page = urllib.request.urlopen('http://www.2literal.com')
    for line in page: print(str(line, encoding = 'utf_8'), end = '')


#### Tuple ( ):
- IMMUTABLE object:  can't insert, append or delete anything in it
- faster than a list

- To create:  t = 1,2,3,4,5
- Comma is required to create a tuple, e.g., one_item_tuple = 1,
- some methods:  len(t), min(t), max(t), .count(t), .index(t)


#### List [ ]:
- mutable
- To create:  use square brackets -- x = [1,2,3,4,5]  
- some methods:  see tuple, .append(x), .extend(x), x.pop(), x.pop(0), .remove(x), .insert(index, x)

- can do replacements, like this:
    list[27:43:3] = (99,99,99,99,99,99)
    (where 27 is beginning index; 43 is end index; 3 is step)
    

#### Dictionary { }:
- an associative array
- mutable

- To create:  d = dict(one = 1, two = 2, three = 3, four = 'four')

- can add one dictionary to another using:  **var_name

- methods:  .get(), .items(), del, .pop(), d['one'] (NOTE: need parens around key!)

- some loops:

        for k, v in d.items(): print (k, v)

        for k in d:
            print(k, d[k])
        
        will print:
        two 2
        one 1
        three 3

- in NO particular order
- to order:

        for k in sorted(d.keys()):
            print(k, d[k])
        
        will print:
        one 1
        three 3
        two 2
        (sorted by key)

#### Loops

- .readlines() is an iterator function
s = 'this is a string'
for line in s.readlines():
    print(line, end='')

- with "enumerate":
s = 'this is a string'
for i, c in enumerate(s):  # index, value (in this case: character)
    if c == 's':
        print('index {} is an s'.format(i)) # .format(i) is necessary because i is not a string
    
- to skip over something and continue with loop
s = 'this is a string'
for c in s:
    if c == 's': continue
    print(c, end='')
    
- or use "break" in the above to break out of the loop entirely

- WHILE:
s = 'this is a string'
i = 0
while( i < len(s)):
    print(s[i], end='')
    i += 1  # no i++

#### Slices & Ranges

s = 'this is a string'
s[0:5]  # will print out "this "

for i in range(0,8): print(i)  # will print "01234567"

#### Regular Expressions

import re

f = open('C:\\Users\\kt\\Desktop\\lynda\\python\\raven.txt')
for line in f:
    if re.search('(Len|Neverm)ore', line):
        print(line, end='')

- using object 'match':

f = open('C:\\Users\\kt\\Desktop\\lynda\\python\\raven.txt')
for line in f:
    match = re.search('(Len|Neverm)ore', line)
    if match:
        print(match.group())

- search and replace:

f = open('C:\\Users\\kt\\Desktop\\lynda\\python\\raven.txt')
for line in f:
    print(re.sub('(Len|Neverm)ore', '###', line), end='')

- more efficient to use a pattern so that the regex is not compiled with every loop
- also can use things like IGNORECASE
f = open('C:\\Users\\kt\\Desktop\\lynda\\python\\raven.txt')
pattern = re.compile('(Len|Neverm)ore', re.IGNORECASE)
for line in f:
    if re.search(pattern, line):
        print(pattern.sub('###', line), end='')

#### Strings

- immutable
- all strings are objects, even in their raw form (not assigned to a variable)

- SEE:  5.6.1 String Methods -- https://docs.python.org/3.1/library/stdtypes.html#string-methods
- some functions are:
    .capitalize()
    .upper()
    .lower()
    .swapcase()
    .find('word')  # will integer return position of 'word'
    .replace('word', 'new word')
    .strip()       # by default will strip whitespace
    .rstrip()      # removes whitespace from end of string
    .rstrip('\n')  # just removes newline
    .isalnum()     # checks if string contains only alphanumeric characters (no whitespace)
    .isalpha()     # checks if string contains only alphabetic characters (no whitespace)
    .isdigit()     # digits
    .isprintable()

    .format()      # allows you to put placeholders in a string -- {} -- 'This is a number: {}'.format(42)
                   # can also use this with dictionaries

    .split()       # splits on whitespace and deletes all whitespace; puts items into a LIST
    .split('char') # splits on character but does NOT delete whitespace; puts items into a LIST
    ':'.join(words)   # joins "words" string items from list with ":" between them


- escape characters:  
\n = new line
r  = raw  e.g., r"This is a \nstring" - will print out "\n"; useful with regular expressions
{} = supplies a value  e.g.:
     n = 42
     s = "This is a {} string!".format(n)
to span lines, use: '''  or """  e.g.,
s = '''\  <- the slash here escapes the newline on this line
This is a string
that spans more 
than one line
'''
NOTE:  There will be line breaks inserted at the end of each of these lines
```
